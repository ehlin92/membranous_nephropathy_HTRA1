```python
import pandas as pd
import numpy as np
```

```python
df_autoimmune = pd.read_csv('../data/automimmune_profiling.txt',sep='\t',index_col='Uniprot')
df_autoimmune['comparison_ID'] = df_autoimmune['Gene']
df_autoimmune.index=df_autoimmune['comparison_ID']
df_autoimmune.head()
```

```python
df_podocyte = pd.read_csv('../data/podocyte_proteins.csv',sep=',',index_col='Accession')
df_podocyte['comparison_ID']=df_podocyte['Description'].apply(lambda x: x.split('[')[1].split('_')[0])
df_podocyte.index=df_podocyte['comparison_ID']
df_podocyte.head()
```

```python
df_exosome = pd.read_csv('../data/urinary_podocyte_exosomes.csv',sep=',')
df_exosome['comparison_ID'] = df_exosome['ID'].apply(lambda x: x.split('_')[0])
df_exosome.index=df_exosome['comparison_ID']
df_exosome.head()
```

```python
index_tmp = pd.Index.intersection(df_exosome.index,df_podocyte.index)
index_shared = pd.Index.intersection(index_tmp,df_autoimmune.index)
df_prioritized=df_autoimmune.loc[index_shared.unique(),:].dropna().sort_values(by='AP0238_A',ascending=False)
df_prioritized.to_excel('../results/df_prioritized_proteins.xlsx')
df_prioritized.head(30)
```
